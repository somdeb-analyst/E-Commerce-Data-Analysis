## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

# Calculation of the total revenue for the E-Commerce company

```python
total_revenue=0               # we define a variable total_revenue and assign an initial value of 0 before starting computation
for i in df['discounted_price']: # we are running a for loop so that all the orders are taken into consideration
    if(i>600):                  # imposing 1st condition on orders priced more than 600
        revenue=0.25*i           #calculating revenue
        total_revenue+=revenue    #updating the value of the variable revenue
    elif(i>350 & i<=600):         #imposing 2nd condition on orders priced more than 350, less than or equal to 600
        revenue=0.15*i            # calculating revenue
        total_revenue+=revenue   #updating the value of the variable revenue
    elif(i>100 & i<=350):          #imposing 3rd condition on orders priced more than 100, less than or equal to 350
        revenue=0.1*i              # calculating revenue
        total_revenue+=revenue    #updating the value of the variable revenue
    else:                           #imposing last condition on orders priced less than or equal to 100
        revenue=0.05*i              # calculating revenue
        total_revenue+=revenue        #updating the value of the variable revenue
print('The total reveue is',np.round(total_revenue,2))   #printing the final revenue rounded upto 2 places after decimal
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


### Ans 6

# Step by step solution of qs no 6 is as follows:

# We first execute a code to add an extra column to the dataframe named company_revenue. This column will contain the revenues generated by the ECom company from each order.

```python
company_revenue=[]
for i in df['discounted_price']:
    if(i>600):
        revenue=0.25*i
        company_revenue.append(revenue)
    elif(i>350 & i<=600):
        revenue=0.15*i
        company_revenue.append(revenue)
    elif(i>100 & i<=350):
        revenue=0.1*i
        company_revenue.append(revenue)
    else:
        revenue=0.05*i
        company_revenue.append(revenue)
df['company_revenue']=pd.Series(company_revenue)
```

# Let us check the dataframe with this new column now.

```python
df.head()
```

# Now the brand revenue = discounted price - company revenue. Therefore if we can introduce a new column brand_revenue into the dataframe where each record will be the difference of the corresponding records in the discounted price column and the company revenue column our brand revenue for each brand will be prepared based on each order placed.

```python
df['brand_revenue']=df['discounted_price']-df['company_revenue']
```

# Let us check the dataframe again.

```python
df.head()
```

# Total Brand Revenue is the sum of the brand revenues of all the products. It can be computed as follows:

```python
total_brand_revenue = round(df['brand_revenue'].sum(),2)
print(total_brand_revenue)
```

# We can list the top 10 Brands having maximum revenue in descending order as follows:

```python
df.groupby('Brand')['brand_revenue'].sum().sort_values(ascending=False).head(10)
```
